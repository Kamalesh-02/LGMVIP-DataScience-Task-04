## Image to Pencil Sketch in Python

### Import the openCV Library
The code begins by importing the necessary libraries, namely cv2 for image processing and matplotlib.pyplot for displaying images.

```python
import cv2
import matplotlib.pyplot as plt
```

### Read the image

```python
image = cv2.imread("C:/Users/kamalesh k b/Downloads/Rdj.jpg")
```

### From openCV BGR to Matplolib RGB
The code converts the original image (stored in the image variable) from the BGR color space to RGB using cv2.cvtColor(). Then, it displays the RGB image using plt.imshow() and shows it with the title "ORIGINAL IMAGE" using plt.title(). The plt.axis(False) command removes the axis ticks, and plt.show() displays the image.

```python
RGB_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.title("ORIGINAL IMAGE")
plt.axis(False)
plt.show()
```

### Original Image to Grey Scale
The code converts the original image to grayscale using cv2.cvtColor() with the cv2.COLOR_BGR2GRAY conversion flag. It displays the grayscale image, adds a title "GRAYSCALE IMAGE," removes the axis ticks, and shows the image.

```python
grey_filter =cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(grey_filter)
plt.title("GRAYSCALE IMAGE")
plt.axis(False)
plt.show()
```

### Invert the Gray scaled Image
The code uses cv2.bitwise_not() to create a negative image by inverting the grayscale image. It then displays the negative image, adds a title "NEGATIVE IMAGE," removes the axis ticks, and shows the image.

```python
invert = cv2.bitwise_not(grey_filter)
plt.imshow(invert)
plt.title("NEGATIVE IMAGE")
plt.axis(False)
plt.show()
```

### Blur the Inverted Image
The code applies a Gaussian blur effect to the inverted image using cv2.GaussianBlur(). The parameters (111, 111) specify the kernel size, and 0 indicates the standard deviation of the blur. It displays the blurred image, adds a title "BLUR IMAGE," removes the axis ticks, and shows the image.

```python
blur = cv2.GaussianBlur(invert,(111,111),0)
plt.imshow(blur)
plt.title("BLUR IMAGE")
plt.axis(False)
plt.show()
```

### Invert the Blured Image
The code applies the cv2.bitwise_not() function again to invert the blurred image. It then displays the inverted blurred image, adds a title "INVERTED BLUR IMAGE," removes the axis ticks, and shows the image.

```python
invertblur = cv2.bitwise_not(blur)
plt.imshow(invertblur)
plt.title("INVERTED BLUR IMAGE")
plt.axis(False)
plt.show()
```

### Sketch the Inverted Blur Images
The code uses cv2.divide() to create a sketch-like image by dividing the grayscale image by the inverted blurred image. The parameter scale=256.0 is used to scale the result. It displays the sketched image, adds a title "SKETCHED IMAGE," removes the axis ticks, and shows the image.

```python
sketch_filter = cv2.divide(grey_filter, invertblur, scale=256.0)
plt.imshow(sketch_filter)
plt.title("SKETCHED IMAGE")
plt.axis(False)
plt.show()
```

### Original vs Pencil Sketch
The code creates a figure with a size of (14, 8) using plt.figure(figsize=(14, 8)). It then creates subplots using plt.subplot() to display the original image and sketch side by side. The first subplot displays the original image with the title "Original image," and the second subplot displays the sketch with the title "Pencil Sketch." The plt.axis('off') command removes the axis ticks for the original image, and cv2.cvtColor() is used to convert the sketch image to RGB format before displaying it. Finally, the plt.show() command shows the figure with both images.

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size = 18)
plt.imshow(RGB_img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Pencil Sketch', size = 18)
RGB_sketch = cv2.cvtColor(sketch_filter, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_sketch)
plt.axis(False)
plt.show()
```
